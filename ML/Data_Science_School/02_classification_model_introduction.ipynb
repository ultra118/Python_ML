## 1개의 독립변수를 가진 가상 데이터

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt

X, y = make_classification(n_features=1, n_informative=1,
                           n_redundant=0, n_clusters_per_class=1, random_state=4)
plt.scatter(X, y, marker='o', c=y,
            s=100, edgecolor="k", linewidth=2)

plt.xlabel("$X$")
plt.ylabel("$y$")
plt.show()
```

## make_blobs
- 등방성 가우시안 정규분포를 이용해 가상 데이터를 생성
- 등방성 : 모든 방향으로 같은 성질을 가진다는 뜻
- 인수
    - n_samples : 표본 데이터의 수
    - n_features : 독립 변수의 수
    - centers : 생성할 클러스터의 수 혹은 중심
    - cluster_std : 클러스터의 표준 편차
    - cetner_box : 생성할 클러스터의 바운딩 박스
- 반환값
    - 독립변수 x : [n_samples, n-features] 크기의 배열
    - 종속변수 y : [n_samples] 크기의 배열

### 세 계의 클러스터를 가진 가상 데이터

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_features=2, centers=3)
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, s=100, edgecolors="k", linewidths=2)
plt.xlabel("$X_1$")
plt.ylabel("$X_2$")
plt.show()
```

#  분류 모형
- 독립 변수 값이 주어졌을 때 그 독립 변수 값과 가장 연관성이 큰 종속변수 카테고리(클래스)를 계산하는 문제
- 분류 문제를 푸는 방법
    - 판별함수 모형
        - 주어진 데이터를 카테고리에 따라 서로 다른 영역으로 나누는 경계면(decision boundary)을 찾아낸 다음 경계면으로부터 주어진 데이터가 어느 위치에 있는지를 계산하는 판별함수(discriminant function)를 이용
    - 확률적 모형
        - 주어진 데이터에 대해 각 카테고리 혹은 클래스가 정답일 조건부 확률을 계산
        - 조건부 확률
            - 확률적 판별(discriminative) : 직접 조건부확률 함수를 추정
                - 
            - 확률적 생성(generative) : 베이즈 정리 사용
                - 각 클래스 별 특징 데이터의 확률분포를 추정
                - 그리고 베이즈 정리를 사용하여 계산
                - 따라서 새로운 가상의 특징 데이터를 생성해내거나 특징 데이터만으로도 아웃라이어를 판단할 수 있음
                - 하지만 클래스가 많을 경우 모든 클래스에 대해 추정하는 것은 많은 데이터를 필요로 할 뿐더러 쓸데없는 계산량을 너무 많이 필요로 함
            

## QDA(Quadratic Discriminat Analysis)
- 조건부확률기반 생성(generative)모형

```python
X = np.array([[-1,-1], [-2,-1], [-3,-2], [1,1], [2,1],[3,2]])
y = np.array([0,0,0,1,1,1])
plt.scatter(X.T[0], X.T[1], c=y, s=100, edgecolor='k', linewidth =2)
plt.show()
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
model = QuadraticDiscriminantAnalysis().fit(X, y)
```

```python
# (0,0)이 주어졌을떄 0에속하는지 1에속하는지 조건부확률
x = [[0, 0]]
p = model.predict_proba(x)[0]

plt.subplot(211)
plt.scatter(X.T[0], X.T[1], c=y, s=100, edgecolor='k', linewidth=2)
plt.scatter(x[0][0], x[0][1], c='r', s=100,
            edgecolor='k', marker='x', linewidth=5)

plt.subplot(212)
plt.bar(model.classes_, p)

plt.gca().xaxis.grid(False)
plt.xticks(model.classes_, ["$P(y=0|x_{test})$", "$P(y=1|x_{test})$"])
plt.tight_layout()
plt.show()
```

```python
x = [[-0.2, -0.1]]
p = model.predict_proba(x)[0]

plt.subplot(211)
plt.scatter(X.T[0], X.T[1], c=y, s=100, edgecolor='k', linewidth=2)
plt.scatter(x[0][0], x[0][1], c='r', s=100,
            marker='x', edgecolor='k', linewidth=5)

plt.subplot(212)
plt.bar(model.classes_, p)

plt.gca().xaxis.grid(False)
plt.xticks(model.classes_, ["$P(y=0|x_{test})$", "$P(y=1|x_{test})$"])
plt.tight_layout()
plt.show()
```

```python
x = [[0.2, 0.2]]
p = model.predict_proba(x)[0]

plt.subplot(211)
plt.scatter(X.T[0], X.T[1], c=y, s=100, edgecolor='k', linewidth=2)
plt.scatter(x[0][0], x[0][1], c='r', s=100,
            marker='x', edgecolor='k', linewidth=5)

plt.subplot(212)
plt.bar(model.classes_, p)

plt.gca().xaxis.grid(False)
plt.xticks(model.classes_, ["$P(y=0|x_{test})$", "$P(y=1|x_{test})$"])
plt.tight_layout()
plt.show()
```

## 나이브 베이지안 모형
- 조건부확률 기반 생성 모형의 장점 중 하나는 클래스가 3개 이상인 경우에도 바로 적용할 수 있다는 점

```python
from sklearn.datasets import fetch_20newsgroups
# 텍스트 데이터를 정해진 크기의 실수벡터로 반환하는 전처리기
from sklearn.feature_extraction.text import TfidfVectorizer
# 나이즈 베이즈 방법으로 분류문제를 예측
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline

news = fetch_20newsgroups(subset="all")
model = Pipeline([('vect', TfidfVectorizer(stop_words="english")),
                  ('nb', MultinomialNB()), ])
model.fit(news.data, news.target)

n = 1
x = news.data[n:n+1]
y = model.predict(x)[0]
print(x[0])
print("=" * 80)
print("실제 클래스:", news.target_names[news.target[n]])
print("예측 클래스:", news.target_names[y])
```

### 나이브 베이지안 모형으로 특정한 텍스트 데이터가 20개의 카테고리에 속할 확률을 각각 계산


```python
plt.subplot(211)
# 확률 값
plt.bar(model.classes_, model.predict_proba(x)[0])
plt.xlim(-1, 20)
plt.gca().xaxis.grid(False)
plt.xticks(model.classes_)
plt.subplot(212)
# 로그 변환
plt.bar(model.classes_, model.predict_log_proba(x)[0])
plt.xlim(-1, 20)
plt.gca().xaxis.grid(False)
plt.xticks(model.classes_)
plt.show()
```

## 확률적 판별 모형(probabilistic discriminative model)
- 조건부 확률이 x에 대한 함수 f(x)로 표시될 수 있다고 가정하고 그 함수를 직접 찾아내는 방법
- 단 f(x)는 0보다 같거나크고 1보다 같거나 작다는 조건을 만족해야 함

### 로지스틱 회귀 모형

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression

X0, y = make_classification(n_features=1, n_redundant=0, n_informative=1, n_clusters_per_class=1, random_state=4)

model = LogisticRegression().fit(X0, y)
xx = np.linspace(-3, 3, 100)
XX = xx[:, np.newaxis]
prob = model.predict_proba(XX)[:, 1]

x_test = [[-0.2]]

plt.subplot(211)
plt.plot(xx, prob)
plt.scatter(X0, y, marker='o', c=y, s=100, edgecolor='k', linewidth=2)
plt.scatter(x_test[0], model.predict_proba(x_test)[0][1:], marker='x', s=500, c='r', lw=5)
plt.xlim(-3, 3)
plt.ylim(-.2, 1.2)
plt.legend(["$P(y=1|x_{test})$"])
plt.subplot(212)
plt.bar(model.classes_, model.predict_proba(x_test)[0])
plt.xlim(-1, 2)
plt.gca().xaxis.grid(False)
plt.xticks(model.classes_, ["$P(y=0|x_{test})$", "$P(y=1|x_{test})$"])

plt.tight_layout()
plt.show()
```

### 판별함수 기반 모형
- 동일한 클래스가 모여 있는 영역과 그 영역을 나누는 경계면(boundary plane)을 정의
- 경계면으로부터의 거리를 계산하는 f(x) 형태의 함수인 판별함수(discriminant function)로 정의
