# 분류용 가상 데이터 생성
- make_classification
    - n_samples : 표본 데이터 수, default 100
    - n_features : 독립 변수의 수, default 20
    - n_informative : 독립 변수 중 종속 변수와 상관관계가 있는 성분의 수, default 2
    - n_redundant : 독립 변수 중 다른 독립 변수의 선형 조합으로 나타나는 성분의 수, default 2
    - n_repeated : 독립 변수 중 단순 중복된 성분의 수, default 0
    - n_classes : 종속 변수의 클래스 수, default 2
    - n_clusters_per_class : 클래스 당 클러스터의 수, default 2
    - weight : 각 클래스에 할당된 표본 수
    - random_state : 난수 발생 시드

## 1개의 독립변수를 가진 가상 데이터

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt

X, y = make_classification(n_features=1, n_informative=1,
                           n_redundant=0, n_clusters_per_class=1, random_state=4)
plt.scatter(X, y, marker='o', c=y,
            s=100, edgecolor="k", linewidth=2)

plt.xlabel("$X$")
plt.ylabel("$y$")
plt.show()
```

## 2개의 독립변수를 가지고 2개의 클래스를 가지는데이터

```python
# n_informative 변수를 1로 설정, 독립변수는 2개인데 실제로 
# target class와 상관관계가 있는 것은 하나의 독립변수
X, y = make_classification(n_features=2, n_informative=1, n_redundant=0, n_clusters_per_class=1, random_state=4)
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y,
            s=100, edgecolor="k", linewidth=2)

plt.xlabel("$X_1$")
plt.ylabel("$X_2$")
plt.show()
```

### 두개의 독립변수 모두 클래스와 상관관계가 있는 가상 데이터
