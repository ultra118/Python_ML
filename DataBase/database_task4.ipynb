```python
# 가장 많이 count 된 명사들 10개 return
def maxcount_return(file_path, data):
    if '_eng' in file_path:
        count = 0
        print("명사만을 추출")
        for word in data['word']:
            tag_value = pos_tag(word_tokenize(word))
            if len(tag_value) == 1 and tag_value[0][1] == "NN" :
                count += 1
                print(word)
            if count == 10:
                break;
        print("그냥 카운트가 큰 것들순으로만(명사 고려 X)")
        display(data.head(10))
    else:
        display(data.head(10))
    
```

```python
import numpy as np
import pandas as pd
from nltk.tag import pos_tag
from nltk.tokenize import word_tokenize

file_name = ['korea_eng.txt', 'korea_kor.txt', 'unitedstates_eng.txt', 'unitedstates_kor.txt',
             'japan_eng.txt', 'japan_kor.txt', 'china_eng.txt', 'china_kor.txt']
df_list = []
for f in file_name:
    file_path = './data/' + f
    df = pd.read_csv(file_path, sep = '\t', header=None)
    df.columns = ['word', 'count']
    df.sort_values(by='count',ascending=False, inplace=True, axis = 0)
    df_list.append(df)
    print("+++++++++++++++{}의 분석결과++++++++++++++++++++++".format(f[:-4]))
    maxcount_return(file_path, df)
```
