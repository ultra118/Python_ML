```python
# 문제2) Logistic regression
# 대학원 입학자료를 이용하여 특정 조건에 따른 입학여부를 예측하는 프로그램을 TensorFlow를 이용하여 작성한다.
# 제공된 CSV 파일로부터 데이터를 loading한 후 학습과정을 진행하고
# 특정 조건(GRE,GPA,RANK)에 따른 합격여부를 예측한다.

import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler

file_path = "./data/admission/admission.csv"
admission_df = pd.read_csv(file_path, sep=",")
# 결측 값 제거
admission_df = admission_df[["gre","gpa","rank","admit"]]
admission_df = admission_df.dropna(how="any")

data_x = admission_df[["gre","gpa","rank"]]
data_y = admission_df["admit"]
# 범주형 데이터를 one-hot으로
rank_dummie = pd.get_dummies(data_x["rank"], prefix="rank")
data_x = data_x.join(rank_dummie)
# 기존의 rank는 지움
data_x.drop("rank", axis = 1, inplace = True)

# min max 일반화
scaler = MinMaxScaler()
x_data = scaler.fit_transform(data_x.values)
y_data = data_y.values.reshape(-1,1)
print("x_data.shape : {}, y_data.shape : {}".format(x_data.shape, y_data.shape))

x_data_train = x_data[:int((x_data.shape[0])*0.8),]
y_data_train = y_data[:int((x_data.shape[0])*0.8),]
x_data_test = x_data[int((x_data.shape[0])*0.8):,]
y_data_test = y_data[int((x_data.shape[0])*0.8):,]
print("trainX,Yshape : {}, {}, testX,Y.shape : {}, {}"
      .format(x_data_train.shape,y_data_train.shape,x_data_test.shape, y_data_test.shape))

# tensor graph
X = tf.placeholder (shape = [None, 6] , dtype = tf.float32)
Y = tf.placeholder (shape = [None, 1] , dtype = tf.float32)

W = tf.Variable(tf.random_normal([6,1]), dtype = tf.float32, name = "wieght")
b = tf.Variable(tf.random_normal([1]), dtype = tf.float32, name = "bias")

logit = tf.matmul(X,W) + b
H = tf.sigmoid(logit)

cost = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits = logit, labels=Y))

optimzer = tf.train.AdamOptimizer(learning_rate=0.001)
train = optimzer.minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(10000):
    _, cost_value, bias_value, weight_value = sess.run([train, cost, b,W],
                                                       feed_dict={X:x_data_train, Y:y_data_train})
    if step % 2000 == 0:
        print("cost_val : {}, weight_value : {}, bias_value : {}".format(cost_value, weight_value, bias_value))

predict = tf.cast(H > 0.5, dtype = tf.float32)        
correct = tf.equal(predict, Y)
accuracy = tf.reduce_mean(tf.cast(correct, dtype = tf.float32))
print("accuracy : {}".format(sess.run(accuracy, feed_dict = {X:x_data_test, Y:y_data_test})))

# 특정 조건(GRE,GPA,RANK)에 따른 합격여부를 예측한다.
input_x = [900,4.0,1]
# 범주형 입력값을 one_hot으로 바꿔줌
if input_x[2] == 1:
    del input_x[2]
    input_x.extend([1,0,0,0])
elif input_x[2] == 2:
    del input_x[2]
    input_x.extend([0,1,0,0])
elif input_x[2] == 3:
    del input_x[2]
    input_x.extend([0,0,1,0])
else:
    del input_x[2]
    input_x.extend([0,0,0,1])
x_predict = np.array(input_x).reshape(-1,6)
# 위에서 표준화할때 사용된 min_max를 통해 표준화 된 값을 집어 넣음
x_predict = scaler.transform(x_predict)
y_predict = sess.run(H,feed_dict={X:x_predict})
if y_predict > 0.5:
    print("predict : {}, 합격".format(y_predict))
else:
    print("predict : {}, 불합격".format(y_predict))

```
