### magic function(jupyter ntebook에서만 사용 가능)
- %로 시작하는 
- 어떤 형태로 사용하는지 지정하는데
    - %matplotlib inline : 그래프를 만든 후 output영역에 그림을 넣어줌
        - 그래프에 대한 추가적인 조작을 할 수 없음
    - %matplotlib nbagg : 그래프를 만든 후 interactive하게 그래프를 조작할 수 있는 창이 생김
        - 창을 통해서 생성한 그래프를 제어할 수 있음(추가적인 조작이 가능)

##### matplotlib의 기본설정
- paremter를 지정, dic의 key값을 통해 접근, 설정
    - plt.rcParams["axes.grid"]
        - 격자를 표현할지 안할지, True면 격자가나옴
    - plt.rcParams["figure_figsize"]
        - 그래프의 크기

```python
%matplotlib nbagg
# numpy를 이용해서 line graph
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

# matplotlib의 기본설정
# 1. parameter 지정, 안에 배열형태로 key값이 들어감
plt.rcParams["axes.grid"] = False #격자를 표현할지, 안할지, True면 격자가 나오고 Flase면 안나옴
plt.rcParams["figure.figsize"] = (8,4) #그래프의 크기

# plotting할 데이터 준비
x_data = range(50)
np.random.seed(0)
# cumsum() : 값들을 누적하면서 더해주는
y_data = np.random.randint(-100, 100, 50).cumsum()
print(y_data)

plt.title("가격변동 추이") # 타이틀
plt.xlabel("기간(일)")
plt.ylabel("가격")
# 기본적인 plot, 점과점을 선으로 잇는
plt.plot(x_data,y_data,"r")
plt.show()
```

##### 한글이 보이도록 설정
- matplotlib.font_manager 사용
    - font_manager.FontProperties(font_path, size) : 어떤 폰트를 어떤 크기로 사용할지 하나의 속성 객체를 만듬
        - font_path와 font_size를 지정해줌, font_path에 사용되는 폰트는 ttf를 선호
- 폰트는 전역으로 설정해주거나, 특정 폰트사용되는 부분에 각기 다른 폰트설정을 할 수 있음

```python
%matplotlib inline
# numpy를 이용해서 line graph
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
# font 처리 하는 module 추가
import matplotlib.font_manager as fm

# matplotlib의 기본설정
# 1. parameter 지정, 안에 배열형태로 key값이 들어감
plt.rcParams["axes.grid"] = False #격자를 표현할지, 안할지, True면 격자가 나오고 Flase면 안나옴
plt.rcParams["figure.figsize"] = (8,4) #그래프의 크기
# 사용할 font의 properties를 생성, ttc보단 ttf형식이 더 좋긴함
font_path = "C:/Windows/Fonts/gulim.ttc"
# 어떤 폰트를 어떤 크기로 사용할지 하나의 속성 객체를 만듬
font_prop = fm.FontProperties(fname = font_path, size = 14)


# plotting할 데이터 준비
x_data = range(50)
np.random.seed(0)
# cumsum() : 값들을 누적하면서 더해주는
y_data = np.random.randint(-100, 100, 50).cumsum()
print(y_data)
# font 속성 여러개 만들어서 원하는 위치에 원하는 속성으로 폰트를 설정해서 나타낼 수 있음
plt.title("가격변동 추이", fontproperties = font_prop) # 타이틀
plt.xlabel("기간(일)", fontproperties = font_prop)
plt.ylabel("가격", fontproperties = font_prop)
# 기본적인 plot, 점과점을 선으로 잇는
plt.plot(x_data,y_data,"r")
plt.show()
```

### matplotlib의 기본설정

```python
%matplotlib inline

import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

plt.rcParams["axes.grid"] = True #격자 ON/OFF
plt.rcParams["figure.figsize"] = (8,4) # 그래프 크기
plt.rcParams["axes.unicode_minus"] = False # "-"글자로 인한 문제발생 제거
plt.rcParams["font.size"] = 16. # 폰트 크기
plt.rcParams["xtick.labelsize"] = 8. # x축 틱에 대한 사이즈
plt.rcParams["ytick.labelsize"] = 8. # y축 틱에 대한 사이즈
plt.rcParams["axes.labelsize"] = 14 # label의 크기
# 전역으로 사용할 font 설정 - ttf파일을 선호
font_path = "C:/Windows/Fonts/NGULIM.ttf"
font_family = fm.FontProperties(fname= font_path).get_name()
plt.rcParams["font.family"] = font_family

# plotting할 데이터 준비
x_data = range(50)
np.random.seed(0)
# cumsum() : 값들을 누적하면서 더해주는
y_data = np.random.randint(-100, 100, 50).cumsum()
print(y_data)

plt.title("가격변동 추이") # 타이틀
plt.xlabel("기간(일)")
plt.ylabel("가격")
# 기본적인 plot, 점과점을 선으로 잇는
plt.plot(x_data,y_data,"r")
plt.show()
```

#### Series인 경우 index를 x축으로 value를 y축으로 사용

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

#plotting할 데이터 준비
s = pd.Series(np.random.randn(10), index = np.arange(0,100,10))

plt.title("가격변동 추이")
plt.xlabel("가격변동 추이")
plt.ylabel("가격")
# Series의 경우 index를 x축으로 values값을 y축으로
plt.plot(s, label="오렌지 가격") #범례 설정
plt.legend() #범례를 표현
plt.show()
```

#### Bar graph (Series)

```python
s = pd.Series(np.random.randint(50,100,(3,)), index = ["국어","영어","수학"])

s.plot(kind = "bar", label="각 과목 성적")
plt.legend()
plt.show()
# barh는 수평으로 보이게 됨
s.plot(kind = "barh", label="각 과목 성적")
plt.show()

```

### 산점도(scatter) : 두개의 독립변수 X,Y의 관계를 알아보기 위해 사용

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

x = np.random.normal(1,1,(100,1))
y = np.random.normal(-2,4,(100,1))

plt.scatter(x,y)
plt.show()
```
