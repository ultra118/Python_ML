# 셀레니움 통해서 페이지 소스 가져오고 bs4로 분석

```python
from selenium import webdriver
```

```python
webdriver_path = './chromedriver.exe'
driver = webdriver.Chrome(webdriver_path)
```

```python
driver.get('https://upbit.com/exchange?code=CRIX.UPBIT.KRW-BTC')
```

# select의 인자로 들어가는 태그와클래스 사이의 공백을 넣어주면 그 노드안의 노드를 접근

```python
import requests
from bs4 import BeautifulSoup
req = driver.page_source
soup=BeautifulSoup(req, 'html.parser')

bit_list = 'table.highlight '
bit_dict = {}
for bit_title, bit_price in zip(soup.select(bit_list + 'td.tit'), soup.select(bit_list + 'td.price')):
    # print(bit_title.text, bit_price.text)
    bit_dict[bit_title.text] = bit_price.text
```

```python
import pandas as pd

bit_dataframe = pd.DataFrame(bit_dict.values(), index=bit_dict.keys())
```
