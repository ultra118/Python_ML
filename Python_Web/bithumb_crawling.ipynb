```python
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
webdriver_path = './chromedriver.exe'
driver = webdriver.Chrome(webdriver_path)

driver.get('https://www.bithumb.com/')
driver.implicitly_wait(2)
```

```python
req = driver.page_source
soup=BeautifulSoup(req, 'html.parser')

col_list = []
for col in soup.select('thead th')[:5]:
    col_list.append(col.text.replace('\n', '').split(' ')[0])
```

```python
def col_match(selected_tag):
    # 코인
    col1 = selected_tag.find_all('td')[0].text.split(' ')[0]
    # 실시간
    col2 = selected_tag.find_all('td')[1].text.split(' ')[0]
    # 변동률
    col3 = selected_tag.find_all('td')[2].text.split(' ')[0]
    # 거래금액
    col4 = selected_tag.find_all('td')[3].text.split(' ')[1]
    # 시가총액
    col5 = selected_tag.find_all('td')[4].text
    row_list = [col1, col2, col3, col4, col5]
    return row_list
    
```

```python
data_list = []
for bit_list in soup.select('tbody.coin_list tr'):
    data_list.append(col_match(bit_list))
    
```

```python
bit_df = pd.DataFrame(data_list, columns=col_list)
```

```python
bit_df
```

```python
with open('./bithumb_info.txt', 'w', encoding='utf-8') as f:
    f.write('\t'.join(col_list) + '\n')
    for bit_list in bit_df.values:
        for bit in bit_list:
            bit = str(bit)
            f.write(bit + '\t')
        f.write('\n')
        
```

```python
def convert_money(df):
    for col in df.columns:
        df[col] = df[col].apply(lambda x:x.replace(',',''))
    return df
```

```python
bit_df_convert = convert_money(bit_df)
```
