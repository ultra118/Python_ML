6. 제일먼저 git repsitory를 생성
   git repository = git directory = local repository
   => git init : 현재 폴더안에 local repository를 생성(.git 폴더로 생성)
	- 내부적으로 저장되는 형태가 존재하는데 관리할 파일을 효과적으로 저장하기 위해 구조를 갖춘 폴더를 내부적으로 만들어 줌

   => git status : 현재 어떤 branch에서 작업을 하고 있는지, commit 여부, 
      git이 추적하고 있는지 없는지(관리가 되는지 안되는지)의 파일들의 상태들을 보여줌

   => git add readme.txt : readme.txt를 git의 관리대상으로 등록

   => git commit readme.txt -m "initial commit": 추적하고 있는 파일을 저장 -m을 붙여서 메세지를 남겨둠

7. branch
   => git branch ( branch 목록을 조회)
   => git branch hotfix ( hotfix라는 branch를 만듬)

8. checkout
   => git check out hotfix : 특정 branch로 이동

9. merge
   => git merge hotfix : hotfix에 있는걸 master에 합침

- git이 프로젝트의 모든파일을 관리할 필요는 없음
    - 관리가 필요없는 파일들은 git이 관리하지 않도록 특정 파일들을 이용해서 설정할 수 있음
        - .gitignore
            - 이 파일은 직접 만들어줘야 함
            - touch .gitignore
            - 해당 내용은 https://www.gitignore.io/ 을통해 설정해줌 
                - 설정된 것들은 빼고 git이 관리하도록 해줌
- git clone : 원격 repository를 local로 복사할 수 있음

- 두개의 repository를 연결
    - git remote add origin https://github.com/ultra118/DeepLearning.git
    - origin이라는 별명으로 현재 repository에 뒤의 https://github.com/ultra118/DeepLearning.git repository를 추가

    - git remote -v라는 명령어를 연결 되어있는지 확인
    - push는 데이터를 밀때(서버쪽에 보낼 때)
    - fetch는 서버에 있는 데이터를 땡겨올때

- git push [어디로 보낼지] [어떤 브렌치를]
- git push origin master
- git push origin --all => 내가 가진 모든 브렌치를 origin에 밀어줌

- git remote update : update만 할 뿐

- git branch -r : 원격 repository의 branch들을 보여줌
- git checkout -t origin/master : orgin/master를 땡겨옴
